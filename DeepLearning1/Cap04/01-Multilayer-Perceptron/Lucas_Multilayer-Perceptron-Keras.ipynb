```python
from keras.models import Sequential 
from keras.layers import Dense 
from sklearn.model_selection import train_test_split 
import numpy
```

```python
# Random Seed 
seed = 7 
numpy.random.seed(seed)
```

```python
# Carregando o dataset 
dataset = numpy.loadtxt('./data.csv', delimiter=',')
```

```python
dataset
```

```python
dataset.shape
```

```python
# Split between X and y
X = dataset[:,0:8]
y = dataset[:,8]
```

```python
# Split between train and test data
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state = seed)
```

```python
print(X_train.shape)
print(y_train.shape)
```

```python
# Create Model
model = Sequential()
model.add(Dense(12, input_dim=8, kernel_initializer='uniform', activation='relu'))
model.add(Dense(8, kernel_initializer='uniform', activation='relu'))
model.add(Dense(1, kernel_initializer='uniform', activation='sigmoid'))
```

```python
model.summary()
```

```python
# Compilando o modelo
model.compile(loss='binary_crossentropy', optimizer='adam',  metrics=[
        'MeanSquaredError',
        'AUC',
    ])
```

```python
# Treinando o modelo
model.fit(X_train, y_train, epochs=150, batch_size=10)
```

```python
# Avaliando o modelo com dados de teste 
loss, accuracy = model.evaluate(X_test, y_test)
print('\nLoss: %.2f, Acuracia: %.2f%%' % (loss, accuracy*100))
```

```python
predictions = model.predict(X)
predictions
```

```python
# Ajustar previsoes e imprimir o resultado final
rounded = [round(x[0]) for x in predictions]
print(rounded)
```

```python
accuracy = numpy.mean(rounded==y)
print('Acuracia das previsoes: %.2f%%' % (accuracy*100))
```
